```python
# importing libraries 
import pandas as pd
import numpy as np
```

```python
#loading dataset
df=pd.read_csv(r"C:\Users\mohan reddy\OneDrive\Desktop\self project\Financial_Data_Analysis.csv")
df
```

```python
df.shape
```

```python
#checking null values
df.isnull().sum()
```

```python
#replacing null values with zero
df=df.fillna(0)
df
```

```python
#after removing null values
df.isnull().sum()
```

```python
#checking duplicates
duplicates=df.duplicated().sum()
print(duplicates)
```

```python
# drop the duplicates
df=df.drop_duplicates()
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#converting date column
df["Transaction_Date"]=pd.to_datetime(df["Transaction_Date"])
```

```python
#adding transaction year,month,day
df["year"]=df["Transaction_Date"].dt.year
df["month"]=df["Transaction_Date"].dt.month
df["day"]=df["Transaction_Date"].dt.day
df
```

```python
#add age group
df["age_group"]=pd.cut(df["Age"],
                      bins=[0, 25, 40, 60, 100],
                      labels=["young", "adult", "middle_age", "senior"])
df
```

```python
# add amount category
df["amount_levl"]=pd.cut(df["Amount"],
                        bins=[-1, 10000, 50000, 100000, 2000000],
                        labels=["low", "medium", "high", "very high"])
df
```

```python
#Add Transaction Flag (Debit / Credit)
df['Txn_Flow'] = np.where(df['Transaction_Type'].isin(["Withdrawal", "EMI"]), 
                          "Debit", "Credit")
df
```

```python
#Add Region_Encoded (Label Encoding)
df['Region_Encoded'] = df['Region'].astype('category').cat.codes
df
```

```python
#Total Transaction Amount by Region
region_amount = df.groupby('Region')['Amount'].sum()
print(region_amount)
```

```python
#Count of Loan Status
print(df['Loan_Status'].value_counts())
```

```python
#Average Amount by Account Type
avg_amt = df.groupby('Account_Type')['Amount'].mean()
print(avg_amt)
```

```python
#Monthly Transaction Count
monthly_txn = df.groupby(['year', 'month']).size()
monthly_txn
```

```python
#Top 10 Highest Transactions
top10=df.nlargest(10, "Amount")
top10
```
